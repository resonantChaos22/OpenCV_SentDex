VIDEO 2

Capturing Video through WebCam

```python
cap = cv2.VideoCapture(0)

#defining codec for writing video. XVID works for all os
fourcc = cv2.VideoWriter_fourcc(*'XVID')

#out is the writer
out = cv2.VideoWriter('output.avi', fourcc, 20.0, (640, 480))

while True:
    ret, frame = cap.read()
    
    #converting to grayscale
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    
    #showing the frames that keep changing due to the loop
    cv2.imshow('frame', frame)
    cv2.imshow('gray', gray)
    
    #writimg the frame in the output.avi using out.....#Only colored frame can be stored
    out.write(frame)
    
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

Video 3

drawing through opencv


```python
img = cv2.imread('Pen.jpeg', cv2.IMREAD_COLOR)
```

```python
#cv2.line(image_obj, start_coordinate, end_coordinate, color, line_width)
cv2.line(img, (0, 0), (100, 200), (255, 0, 0), 5)

#cv2.rectangle(image_obj, start_coordinate(top left coordinate), end_coordinate(bottom right coordinate, color, line_width)
cv2.rectangle(img, (100, 200), (200, 400), (0, 255, 0), 5)

#cv2.circle(image_obj, center_coordinate, radius, color, line_width)
cv2.circle(img, (200, 400), 100, (0, 0, 255) ,-1)
#linewidth=-1 fills the inside of the shape

#defining points
pts = np.array([[600, 200], [600, 1000], [10, 900]], np.int32)
#For the polylines to work the array has to be 2*1

#cv2.polylines(image_obj, [numpy_array_of_points(2*1)], <Whether or not the first a nd last point are to be joined>, Color, line_width) )
cv2.polylines(img, [pts], True, (255, 255, 0), 5)

font = cv2.FONT_HERSHEY_COMPLEX
#cv2.putText(img, Text, start_coordinate, font, size, color, line-width )
cv2.putText(img, "Pen", (500, 600), font, 3, (200, 255, 0), 1)
```

```python
cv2.imshow('img',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
roi = img1[0:rows, 0:columns]
img2gray = cv2.cvtColor(img4, cv2.COLOR_BGR2GRAY)
```

```python
#cv2.threshold(image_obj, val1, val2, type)
#val1 = value after the value will be converted to val2
ret, mask = cv2.threshold(img2gray, 50, 255, cv2.THRESH_BINARY_INV)
print(ret)
cv2.imshow('img', mask)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
mask_inv = cv2.bitwise_not(mask)
cv2.imshow('img', mask_inv)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
img1_bg = cv2.bitwise_and(roi, roi, mask=mask_inv) #values are 0 where the plaque is to be placed
img1_bg1 = cv2.bitwise_and(roi, roi, mask=mask)
img2_fg = cv2.bitwise_and(img4, img4, mask=mask)#values are 0 where palque is not to be place 
img2_fg1 = cv2.bitwise_and(img4, img4, mask=mask_inv)
iii = cv2.bitwise_and(roi, roi, mask=img2gray)
cv2.imshow('ii', iii)
cv2.imshow('img1', img1_bg)
cv2.imshow('img2', img2_fg)
cv2.imshow('img11', img1_bg1)
cv2.imshow('img21', img2_fg1)
cv2.waitKey(0)
cv2.destroyAllWindows()

#It is done to get the roi with the plaque
```
